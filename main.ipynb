```sage
R.<x1, x2, x3, x4, x5, x6, y1, y2, y3, y4, y5, y6> = PolynomialRing(QQ)
var_list = R.gens()

f1 = x3*y3+x4*y5
f2 = x5*y3+x6*y5
f3 = x3*y4 + x4*y6 
f4 = x5*y4+x6*y6 

f5 = y3*x3+y4*x5
f6 = y5*x3+y6*x5
f7 = y3*x4 + y4*x6 
f8 = y5*x4+y6*x6 

f9 = y1*x3 + y2 * x5 - x1 
f10 = y1*x4+y2*x6-x2 

F = [f1, f2, f3, f4, f5, f6, f7, f8, f9, f10]

J = jacobian(F, var_list)

print(J)
print(J.rank())

counter = 0 
for i, f in enumerate(F):
    others = F[:i] + F[i+1:]
    Q = R.ideal(others)

    if f in J: 
        counter += 1 
        print(f"{f} is redundant")

print(f"the minimal number of generators is {len(F) - counter}")
```
